```python
import pandas as pd
#chargement du dataset avec encodage specifique
df=pd.read_csv('online_retail.csv',encoding="unicode_escape")
#verification des colonnes obligatoires
print(df.info())
```

```python
df.head()
```

```python
# Identifying missing values
print(df.isnull().sum())
```

```python
df.shape
```

```python
# Suppression des lignes avec CustomerID null
df_clean=df.dropna(subset=['CustomerID'])
df_clean.shape
```

```python
print(df_clean.isnull().sum())
```

```python
#suppression des doublons
df_clean=df_clean.drop_duplicates()
```

```python
df_clean.shape
```

```python
df_clean=df_clean[df_clean['Quantity']>0]
```

```python
df_clean.shape
```

```python
df_clean.to_csv('online_retail_cleaned.csv',index=False)
```
